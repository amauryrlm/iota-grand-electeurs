```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
```

# 2D GRID

# Convergence according to the initial division probability without malicious nodes for different network sizes

```python
def convergence_proba_initiale(Prob,name,size):
    convergence=0
    file_name= "./" + name + "/PROBA" + str(int(Prob*100)) +".csv"
    data=pd.read_csv(file_name,sep=';',header=None)
    data["sum"] = data.sum(axis=1)
    for i in range(0,data.shape[0]):
        if data["sum"][i]==0 or data["sum"][i]==size:
            convergence+=1
    return convergence*100/data.shape[0]
```

```python
P=np.arange(0,1.01,0.01)
GRID_PROBA_NODE49=[]
GRID_PROBA_NODE81=[]
GRID_PROBA_NODE225=[]
GRID_PROBA_NODE529=[]
GRID_PROBA_NODE1024=[]
for i in P:
    GRID_PROBA_NODE49.append(convergence_proba_initiale(i,"PROBA_GRID_NODE49",49))
    GRID_PROBA_NODE81.append(convergence_proba_initiale(i,"PROBA_GRID_NODE81",81))
    GRID_PROBA_NODE225.append(convergence_proba_initiale(i,"PROBA_GRID_NODE225",225))
    GRID_PROBA_NODE529.append(convergence_proba_initiale(i,"PROBA_GRID_NODE529",529))
    GRID_PROBA_NODE1024.append(convergence_proba_initiale(i,"PROBA_GRID_NODE1024",1024))
    
```

```python
ORDER=10

fig = plt.figure(figsize = [10, 7])
ax = fig.add_subplot()

ax.text(0.86, 73,'lambda='+str(4)+'\nQuorum size='+str(10)+'\n\u03C4='+str(0.5)+'\n\u03B2='+str(0.25)+'\nq='+str(0) + '\nROUNDS =' + str(30) , style='italic',
        bbox={ 'pad': 10})
axes = plt.gca()
axes.set_ylim(0, 110)
plt.xlabel("Rate of the opinion 0")
plt.ylabel("Convergence rate")
plt.plot(P, savgol_filter(GRID_PROBA_NODE49, 51, ORDER), label = "49 Nodes")
plt.plot(P, savgol_filter(GRID_PROBA_NODE81, 51, ORDER), label = "81 Nodes")
plt.plot(P, savgol_filter(GRID_PROBA_NODE225, 51, ORDER), label = "225 Nodes")
plt.plot(P, savgol_filter(GRID_PROBA_NODE529, 51, ORDER), label = "529 Nodes")
plt.plot(P, savgol_filter(GRID_PROBA_NODE1024, 51, ORDER), label = "1024 Nodes")
plt.title("Convergence rate according to the initial division of opinions for the grid without malicious nodes")
plt.legend()
plt.show()
```

# Convergence en fonction de la distance et du nombre de noeuds sonds

```python
def convergence_distance_k(distance,k):
    convergence=0
    
    file_name= "./Rsultats dcentraliss/Distance_k_grille_csv/DISTANCE" + str(distance) + "_K" + str(k) +".csv"
    data=pd.read_csv(file_name,sep=';',header=None)
    data["sum"] = data.sum(axis=1)
    for i in range(0,data.shape[0]):
        if data["sum"][i]==0 or data["sum"][i]==N:
            convergence+=1
    return convergence*100/data.shape[0]
```

```python
x= np.arange(1,25,1)
y= np.arange(1,25,1)
xx,yy = np.meshgrid(x,y)
ConvVal=np.zeros(xx.shape)
for i in range(x.size):
    for j in range(y.size):
        ConvVal[i,j]= convergence_distance_k(xx[i,j],yy[i,j])
        
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.set_xlabel('Distance', fontsize=20, rotation=150)
ax.set_ylabel('K', fontsize=20)
ax.text(-5, -5, 80,'N='+str(N)+'\n\u03C4='+str(round(TAU,2))+'\n\u03B2='+str(BETA)+'\nq='+str(q)+'\nTMAX='+str(TMAX), style='italic',
        bbox={ 'pad': 10})
surf= ax.plot_surface(xx,yy,ConvVal,cmap='viridis')
fig.colorbar(surf,shrink=0.5,aspect=5)
plt.title("Convergence en fonction de la distance de la marche alatoire et du nombre de noeuds sonds")
plt.show()

```
