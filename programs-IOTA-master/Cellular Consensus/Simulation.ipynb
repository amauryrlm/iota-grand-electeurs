```python
import csv
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
```

# 2D GRID

# Convergence according to the initial division probability without malicious nodes

### Parameters :
### . n = 49 to 1024
### . Number of Rounds = 20

```python
def convergence_proba_initiale(Prob,file):
    convergence=0
    count = 0
    csvfile = open(file+'/CONV'+str(int(100*Prob+0.5))+'.csv')
    spamreader = csv.reader(csvfile, delimiter=';')
    for row in spamreader:
        count += 1
        del row[-1]
        change = [int(i) for i in row]
        if not(-1 in change) :
            somme = sum(change)
            if somme == 0 or somme == len(change) :
                convergence += 1
    return convergence*100/count

PROB_GRID_1024=[]
PROB_GRID_529=[]
PROB_GRID_225=[]
PROB_GRID_81=[]
PROB_GRID_49=[]
P=np.arange(0,1.01,0.01)
for i in P:
    PROB_GRID_1024.append(convergence_proba_initiale(i,"Simulations_Results/GRID2D/No_Malicious/Conv_without_malicious_initial_division/1024"))
    PROB_GRID_529.append(convergence_proba_initiale(i,"Simulations_Results/GRID2D/No_Malicious/Conv_without_malicious_initial_division/529"))
    PROB_GRID_225.append(convergence_proba_initiale(i,"Simulations_Results/GRID2D/No_Malicious/Conv_without_malicious_initial_division/225"))
    PROB_GRID_81.append(convergence_proba_initiale(i,"Simulations_Results/GRID2D/No_Malicious/Conv_without_malicious_initial_division/81"))
    PROB_GRID_49.append(convergence_proba_initiale(i,"Simulations_Results/GRID2D/No_Malicious/Conv_without_malicious_initial_division/49"))


plt.figure(figsize = [15, 10])
polynomial_order = 10
axes = plt.gca()
axes.set_ylim(0, 110)
plt.xlabel ("Probability $P_0$")
plt.ylabel ("Percentage of convergence")
plt.plot(P,  savgol_filter(PROB_GRID_1024, 51, polynomial_order), label="1024 nodes")
plt.plot(P,  savgol_filter(PROB_GRID_529, 51, polynomial_order), label="529 nodes")
plt.plot(P,  savgol_filter(PROB_GRID_225, 51, polynomial_order), label="225 nodes")
plt.plot(P,  savgol_filter(PROB_GRID_81, 51, polynomial_order), label="81 nodes")
plt.plot(P,  savgol_filter(PROB_GRID_49, 51, polynomial_order), label="49 nodes")
plt.legend()
plt.show()
```
